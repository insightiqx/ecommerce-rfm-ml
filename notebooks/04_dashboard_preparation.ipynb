```python
import pandas as pd

df_sales = pd.read_parquet("../data/processed/superstore_clean.parquet")
df_sales.head()
```

```python
df_sales["Order Date"] = pd.to_datetime(df_sales["Order Date"])
```

```python
# Establecer Order Date como índice temporal
df_sales = df_sales.set_index("Order Date")

sales_by_month = (
    df_sales["TotalPrice"]
    .resample("M")
    .sum()
    .reset_index()
    .rename(columns={"TotalPrice": "sales"})
)

sales_by_month.head()
```

```python
def format_miles(x):
    return f"{int(x):,}".replace(",", ".")

sales_by_month["sales_fmt"] = sales_by_month["sales"].apply(format_miles)

sales_by_month.head()
```

```python
sales_by_month.to_parquet(
    "../data/processed/dashboard_sales_by_month.parquet",
    index=False
)
```

```python
import pandas as pd

df_sales = pd.read_parquet("../data/processed/superstore_clean.parquet")
df_sales["Order Date"] = pd.to_datetime(df_sales["Order Date"])
df_sales.head()
```

```python
# Ventas por región
sales_by_region = (
    df_sales.groupby("Region")["TotalPrice"]
    .sum()
    .reset_index()
    .rename(columns={"TotalPrice": "sales"})
)

# Ventas por segmento
sales_by_segment = (
    df_sales.groupby("Segment")["TotalPrice"]
    .sum()
    .reset_index()
    .rename(columns={"TotalPrice": "sales"})
)

sales_by_region.head(), sales_by_segment.head()
```

```python
sales_by_region["sales_fmt"] = sales_by_region["sales"].apply(format_miles)
sales_by_segment["sales_fmt"] = sales_by_segment["sales"].apply(format_miles)

sales_by_region, sales_by_segment
```

```python
import pandas as pd
import json
import math
import os

# 1️⃣ Cargar datos necesarios
df_sales = pd.read_parquet("../data/processed/superstore_clean.parquet")
df_sales["Order Date"] = pd.to_datetime(df_sales["Order Date"])

model_df_fe = pd.read_parquet("../data/processed/model_df_fe.parquet")

print("Columnas de model_df_fe:")
print(model_df_fe.columns.tolist())

# Aseguramos que churn_pred exista
churn_threshold = 0.5

if "churn_pred" not in model_df_fe.columns:
    model_df_fe["churn_pred"] = (model_df_fe["proba_churn"] >= churn_threshold).astype(int)

# 2️⃣ Funciones de formato
def fmt_int(x):
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return "-"
    return f"{int(round(x)):,}".replace(",", ".")

def fmt_currency(x, symbol="€"):
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return "-"
    
    # Convertir sin decimales y con puntos como separador de miles
    base = f"{int(round(x)):,}".replace(",", ".")
    
    # Devolver string limpio con símbolo euro correcto
    return f"{base} €"

def fmt_percent(x, decimals=1):
    if x is None or (isinstance(x, float) and math.isnan(x)):
        return "-"
    value = round(x * 100, decimals)
    txt = f"{value:,.{decimals}f}"
    txt = txt.replace(",", "X").replace(".", ",").replace("X", ".")
    return f"{txt} %"

# 3️⃣ Calcular KPIs numéricos
total_customers = model_df_fe["Customer ID"].nunique()
total_sales = float(df_sales["TotalPrice"].sum())
avg_ticket = float(
    df_sales.groupby("Order ID")["TotalPrice"].sum().mean()
)
churn_rate = float(model_df_fe["churn_pred"].mean())
avg_proba_churn = float(model_df_fe["proba_churn"].mean())

kpis_numeric = {
    "total_customers": total_customers,
    "total_sales": total_sales,
    "avg_ticket": avg_ticket,
    "churn_rate": churn_rate,
    "avg_proba_churn": avg_proba_churn,
    "churn_threshold": churn_threshold,
}

kpis_formatted = {
    "total_customers": fmt_int(total_customers),
    "total_sales": fmt_currency(total_sales),
    "avg_ticket": fmt_currency(avg_ticket),
    "churn_rate": fmt_percent(churn_rate),
    "avg_proba_churn": fmt_percent(avg_proba_churn),
    "churn_threshold": fmt_percent(churn_threshold),
}

print("\nKPIs numéricos:")
print(kpis_numeric)
print("\nKPIs formateados:")
print(kpis_formatted)

# 4️⃣ Guardar JSON en ../data/processed
output_dir = "../data/processed"
os.makedirs(output_dir, exist_ok=True)

numeric_path = os.path.join(output_dir, "dashboard_kpis_numeric.json")
formatted_path = os.path.join(output_dir, "dashboard_kpis_formatted.json")

with open(numeric_path, "w") as f:
    json.dump(kpis_numeric, f, indent=4)

with open(formatted_path, "w", encoding="utf-8") as f:
    json.dump(kpis_formatted, f, indent=4, ensure_ascii=False)

print("\nArchivos escritos:")
print(" -", numeric_path)
print(" -", formatted_path)

print("\nContenido actual de ../data/processed:")
print(os.listdir(output_dir))
```
