```python
import pandas as pd

file_path = r"C:\Proyectos\ecommerce-rfm-ml\data\raw\Sample - Superstore.csv"

df = pd.read_csv(file_path, encoding="ISO-8859-1")

df.columns
```

```python
df.head()
```

```python
# Eliminar filas sin Customer ID
df = df.dropna(subset=["Customer ID"])

# Asegurarse de que Quantity y Sales son numéricas
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")

# Eliminar filas con Quantity o Sales nulas
df = df.dropna(subset=["Quantity", "Sales"])

# Filtrar cantidades y ventas > 0
df = df[(df["Quantity"] > 0) & (df["Sales"] > 0)]

# Crear TotalPrice (equivalente a Sales para consistencia)
df["TotalPrice"] = df["Sales"]

# Convertir fechas
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```
